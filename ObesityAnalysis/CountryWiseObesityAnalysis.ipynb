```python
data.dropna(inplace=True)
data.drop([2],inplace=True)
data.shape
```

```python
data.reset_index(drop=True,inplace=True)
data.head()
```

```python
data.rename(columns={'Unnamed: 0': 'Country'}, inplace=True)
data.head()
```

```python
data_melt = data.melt(id_vars=['Country'],var_name='Year',value_name='Obesity(%)')
```

```python
data_melt[['Year','Sex']] = data_melt['Year'].str.split(".", expand=True)
data_melt.head()
```

<p>Splitting the data into Male, Female and Both Sexes


```python
data_melt["Sex"] = data_melt["Sex"].map({None:'Both Sexes','1':'Male','2':'Female'})
data_melt[data_melt['Country']=="Albania"]
```

<B>The copy method is used as an error can occur as pandas may return a view and not a copy

```python
male_data = data_melt[data_melt.Sex == 'Male'].copy()
male_data.drop(['Sex'],axis=1,inplace=True)
female_data = data_melt[data_melt.Sex == 'Female'].copy()
female_data.drop(['Sex'],axis=1,inplace=True)
both_data = data_melt[data_melt.Sex == 'Both Sexes'].copy()
both_data.drop(['Sex'],axis=1,inplace=True)
```
