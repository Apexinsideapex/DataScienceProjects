```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
train_data = train_data.drop(['Name','Ticket'], axis=1)
```

```python
train_data.head()
```

```python
train_data.nunique()
```

```python
len(train_data)
```

```python
len(train_data.dropna())
```

```python
train_data['Cabin'].isna().sum()
```

<p>Dropping Cabin as there are too many NaN values

```python
train_data = train_data.drop(['Cabin'],axis=1)
```

```python
train_data[train_data['Pclass']==3].mean()
```

```python
train_data[train_data['Pclass']==2].mean()
```

```python
train_data[train_data['Pclass']==1].mean()
```

<p>Can drop Fare as Passenger class column is indicative of Fare

```python
train_data = train_data.drop(['Fare'], axis=1)
```

```python
train_data.head()
```

```python
train_data['Sex'].replace({'female':0,'male':1},inplace=True)
```

```python
train_data['Embarked'].unique()
```

```python
train_data['Embarked'].replace({'S':0,'C':1,'Q':3,np.nan:0}, inplace=True)
train_data.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.regularizers import l2
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
model = Sequential()
```

```python
X_train,x_test,Y_train,y_test = train_test_split(train_data[['Pclass','Sex','Age','SibSp','Parch','Embarked']], train_data['Survived'])
```

```python
X_train.shape,Y_train.shape
```

```python
model.add(Dense(300,activation='relu',input_dim=6,kernel_regularizer=l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(150,activation='relu',kernel_regularizer=l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(100,activation='relu',kernel_regularizer=l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(50,activation='relu',kernel_regularizer=l2(0.01)))
model.add(Dropout(0.2))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
mean_age=train_data['Age'].mean()
```

```python
train_data.replace({np.nan:abs(mean_age)},inplace=True)
```

```python
model_op = model.fit(X_train,Y_train,epochs=100,verbose=1,validation_data=(x_test,y_test))
```

```python
test_model_op = model.fit(train_data[['Pclass','Sex','Age','SibSp','Parch','Embarked']], train_data['Survived'],epochs=100,verbose=1,validation_data=(x_test,y_test))
```

```python
def loss_func(y_true,y_pred):
    return tf.reduce_mean(y_pred-y_true)+0.00001
```

```python
opt = tf.keras.optimizers.Adam(
    learning_rate=0.2,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
)
```

```python
model.compile(loss=loss_func, optimizer=opt, metrics=['accuracy'])
```

```python
test_model_op = model.fit(train_data[['Pclass','Sex','Age','SibSp','Parch','Embarked']], train_data['Survived'],epochs=100, verbose=0)
```
